```python
import os as os
import pandas as pd
import xml.etree.ElementTree as ET
import numpy as np
import re
from datetime import datetime

# read file
path_parent = os.path.dirname(os.getcwd())
df = pd.read_csv(path_parent+"/Output/Abnormal_Concentration.csv")


# deleting empty rows , alter data , create new column

df=df[df["Patient Age"] != "Not Specified"]

df.loc[df['Patient Age'] == "Children (1 - 13 years old)", 'Patient Age'] = "Children (1 - <13 years old)"
df.loc[df['Patient Age'] == "Children (1-13 uears old)", 'Patient Age'] = "Children (1 - <13 years old)"
df.loc[df['Patient Age'] == "Children (1-13 years old)", 'Patient Age'] = "Children (1 - <13 years old)"
df.loc[df['Patient Age'] == "Children (3 months - 6 years old)", 'Patient Age'] = "Children (3 months - <6 years old)"
df.loc[df['Patient Age'] == "Adult Not Specified", 'Patient Age'] = "Adult (>18 years old)"

df["min_age"] = ""
df["max_age"] = ""
df["min_range"]=""
df["max_range"]=""

# standardise age into numerical format

df.loc[df['Patient Age'] == "Adolescent (13-18 years old)", 'min_age'] = 13
df.loc[df['Patient Age'] == "Adolescent (13-18 years old)", 'max_age'] = 18

df.loc[df['Patient Age'] == "Adult (>18 years old)", 'min_age'] = 18
df.loc[df['Patient Age'] == "Adult (>18 years old)", 'max_age'] = 64

df.loc[df['Patient Age'] == "Adult (60 years old)", 'min_age'] = 60
df.loc[df['Patient Age'] == "Adult (60 years old)", 'max_age'] = 60

df.loc[df['Patient Age'] == "Children (1 - <13 years old)", 'min_age'] = 1
df.loc[df['Patient Age'] == "Children (1 - <13 years old)", 'max_age'] = 13

df.loc[df['Patient Age'] == "Children (3 months - <6 years old)", 'min_age'] = 0.03
df.loc[df['Patient Age'] == "Children (3 months - <6 years old)", 'max_age'] = 6

df.loc[df['Patient Age'] == "Elderly (>65 years old)", 'min_age'] = 65
df.loc[df['Patient Age'] == "Elderly (>65 years old)", 'max_age'] = 99

df.loc[df['Patient Age'] == "Infant (0-1 year old)", 'min_age'] = 0
df.loc[df['Patient Age'] == "Infant (0-1 year old)", 'max_age'] = 1

df.loc[df['Patient Age'] == "Newborn (0-30 days old)", 'min_age'] = 0
df.loc[df['Patient Age'] == "Newborn (0-30 days old)", 'max_age'] = 0.0030

# rename row and drop unneccesary column

df = df.rename(columns={"Unnamed: 0": "id"})

del df['Patient Age']
df = df.drop(df.columns[0], axis=1) 



# get value data

#if alphabet exist
for index, row in df.iterrows():
    temp=""
    value = row["Value(uM)"]
   
    if(re.search('[a-zA-Z]', str(value))):
        row["Value(uM)"]=""
    if(row["Value(uM)"]==""):
        row["min_range"] = ""
        row["min_range"] = ""
# if "<"
# example = <500
    if("<" in row["Value(uM)"]):
        temp = value.split('<')[1]
    try:
        row["min_range"] = 0
        row["max_range"] = float(temp)
    except:
        row["min_range"] = ""
        row["min_range"] = ""
        
    if(">" in row["Value(uM)"]):
        
        temp = value.split('>')[1]
        row["min_range"] = float(temp)
        row["max_range"] = 9999
 

# if "-" exist
# example = 200 -500
    elif("-" in row["Value(uM)"]):
        
        if not("+" in row["Value(uM)"]):
            if not ("(" in row["Value(uM)"]):
               
                try:
                    row["min_range"]=float(value.split('-')[0])
                    row["max_range"]=float(value.split('-')[1])
                except:
                    row["min_range"]=""
                    row["max_range"]=""
# if "-" exist and bracket "()"
# example = 36.0 (13.0-95.0)
            else:
                temp =value.split('-')[0]
                row["min_range"]=float(temp.split('(')[1])
                temp=value.split('-')[1]
                row["max_range"]=float(temp.split(')')[0])
        else:
# if "+/-"
# example = 9.1 +/- 7.2
            if not("(" in row["Value(uM)"]):
                
                temp = value.split('+/-')[0]
                deviation = value.split('+/-')[1]
                try:
                    row["min_range"] = float(temp) - float(deviation)
                    row["max_range"] = float(temp) + float(deviation)
                except:
                    row["min_range"]=""
                    row["max_range"]=""
#  if "+/-" outside ()
# 1.200 +/- 1.2(0-3.2)
            else:
                if("+" in (value.split('(')[0])):

                    
                    row["min_range"]= float((value.split('(')[1]).split('-')[0])
                    row["max_range"]=float(((value.split('(')[1]).split('-')[1]).split(')')[0])

                else:
#  if "+/-" inside ()       
# 0.02 (0.03 +/- 0.01)
                    
                    row["min_range"]=float((value.split('(')[1]).split('+')[0])
                    row["max_range"]=float((value.split('-')[1]).split(')')[0])

    elif("(" in row["Value(uM)"]):
        if("–" in row["Value(uM)"]):
            
            row["min_range"]=float((value.split('–')[0]).split('(')[1])
            row["max_range"]=float((value.split('–')[1]).split(')')[0])



        else:
           
            row["min_range"]=float(value.split('(')[0])-float((value.split('(')[1]).split(')')[0])
            row["max_range"]=float(value.split('(')[0])+float((value.split('(')[1]).split(')')[0])


# delete row with empty range value

df=df[df["Value(uM)"] != ""]
df=df[df["min_range"] != ""]
df=df[df["max_range"] != ""]
df = df.drop(df.columns[2], axis=1) 

# show data
df = df[['Name','Accession','Patient Sex','min_age','max_age','min_range','max_range','Patient Information']]
df.tail(10)

# output into new csv 
# df.to_csv(path_parent+"/Output/hmdb_metabolites.csv",header='False')
```
