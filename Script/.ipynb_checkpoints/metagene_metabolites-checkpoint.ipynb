```python
import os as os
import pandas as pd
import xml.etree.ElementTree as ET
import numpy as np


metagene_array=[]

for i in range(1095) :

# read xml template c
    path_parent = os.path.dirname(os.getcwd())
    file_name = path_parent+"/Output/XML-TemplateC/Disease-{}.xml".format(i)
    tree = ET.parse(file_name)
    root=tree.getroot()
    name=root.find("Name").text
    synonym=root.find("Synonym").text
    
# append into array
    for metabolite in root.find("Metabolites"):
        if(metabolite.find('specimen').text=="plasma" or metabolite.find('specimen').text=="blood" ):
            for row in metabolite:
                
                if(row.tag=="id"):
                    meta_id=row.text
                if(row.tag=="name"):
                    meta_name=row.text
                if(row.tag=="min"):
                    min_range =row.text
                if(row.tag=="max"):
                    max_range=row.text
                if(row.tag=="unit"):
                    unit=row.text
                if(row.tag=="age"):
                    age=row.text
                    
            metagene_array.append([name,synonym,meta_id,meta_name,min_range,max_range,age,unit])


# convert array into df
                
metagene_array = np.array(metagene_array)
df=pd.DataFrame(metagene_array, columns=["name","synonym","meta_id","meta_name","min_range","max_range","age","unit"]) 

# add new column

df["min_age"] = ""
df["max_age"] = ""

# standardise age format YY,MMDD

df=df[df["age"] != "no data"]

df.loc[df['age'] == "1-7 years", 'min_age'] = 1
df.loc[df['age'] == "1-7 years", 'max_age'] = 7

df.loc[df['age'] == "6-10 years", 'min_age'] = 6
df.loc[df['age'] == "6-10 years", 'max_age'] = 10

df.loc[df['age'] == "7-15 years", 'min_age'] = 7
df.loc[df['age'] == "7-15 years", 'max_age'] = 15

df.loc[df['age'] == ">5 years", 'min_age'] = 5
df.loc[df['age'] == ">5 years", 'max_age'] = 99

df.loc[df['age'] == "ALL", 'min_age'] = 0
df.loc[df['age'] == "ALL", 'max_age'] = 99

df.loc[df['age'] == "Adolescence", 'min_age'] = 13
df.loc[df['age'] == "Adolescence", 'max_age'] = 18

df.loc[df['age'] == "Adulthood", 'min_age'] = 18
df.loc[df['age'] == "Adulthood", 'max_age'] = 64

df.loc[df['age'] == "Childhood", 'min_age'] = 1
df.loc[df['age'] == "Childhood", 'max_age'] = 13

df.loc[df['age'] == "Infancy", 'min_age'] = 0
df.loc[df['age'] == "Infancy", 'max_age'] = 2

# 28 days
df.loc[df['age'] == "Neonatal", 'min_age'] = 0
df.loc[df['age'] == "Neonatal", 'max_age'] = 0.0028

df.loc[df['age'] == "adult", 'min_age'] = 18
df.loc[df['age'] == "adult", 'max_age'] = 64

df.loc[df['age'] == "all", 'min_age'] = 0
df.loc[df['age'] == "all", 'max_age'] = 99

df.loc[df['age'] == "child", 'min_age'] = 1
df.loc[df['age'] == "child", 'max_age'] = 13

df.loc[df['age'] == "infancy", 'min_age'] = 0
df.loc[df['age'] == "infancy", 'max_age'] = 2

df.loc[df['age'] == "infant", 'min_age'] = 0
df.loc[df['age'] == "infant", 'max_age'] = 2

df.loc[df['age'] == "newborn", 'min_age'] = 0
df.loc[df['age'] == "newborn", 'max_age'] = 0.0200

# 37 week
# 9 month 1 week
# 9 month 7 days
df.loc[df['age'] == "premature", 'min_age'] = 0
df.loc[df['age'] == "premature", 'max_age'] = 0.0907


# clean non-existent unit data row

df=df[df["unit"] != "no unit"]
df=df[df["unit"] != "not detectable"]
df=df[df["unit"] != "no data"]
df=df[df["unit"] != "not detected"]
df=df[df["unit"] != "%"]
df=df[df["unit"] != "% activity"]
df=df[df["unit"] != "absent"]
df=df[df["min_range"] != ""]
df=df[df["max_range"] != ""]

# replace column name and delete unnecessary column
df=df.rename(columns={"name": "Disease", "synonym": "Disease Synonym", "meta_name": "Name"})
df=df.drop(columns=['age'])

df = df[['Name', 'Disease', 'Disease Synonym','min_age','max_age','min_range','max_range','unit' , 'meta_id']]




# output into csv
df.to_csv(path_parent+"/Output/metagene_metabolites.csv",header='False')

# show df
df.head(100)
```
