```python
import os as os
import pandas as pd

path_parent = os.path.dirname(os.getcwd())

df = pd.read_csv(path_parent+"/Output/Final_hmdb_metagene.csv")
metagene_df = pd.read_csv(path_parent+"/Output/metagene_metabolites.csv")

df=df.rename(columns={"Meta_id": "meta_id"})
merged = pd.merge(df,metagene_df,on='meta_id')
new_df=merged.drop(['Unnamed: 0_x', 'Unnamed: 0_y'], axis=1)

new_df.tail(20)

new_df.to_csv(path_parent+"/Output/merged_with_value_HMDB.csv",header='False')
```
